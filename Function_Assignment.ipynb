**Q1. What is the difference between a function and a method in Python?**
- In Python, functions and methods are both blocks of reusable code that perform specific tasks, but they differ significantly in their context and usage.

*Function*: A function is a standalone block of code that can be defined outside of any class. It can take parameters, perform operations, and return values. Functions promote reusability across different parts of a program.

*Method*: A method is a function that is associated with an object (or class). It is defined within a class and operates on the data contained in that class. Methods require an instance of the class to be invoked and typically manipulate the object's state.

**Example**


```python
#Function Example
def add_numbers(x, y):
    return x + y

result = add_numbers(3, 5)
print(result)# Output: 8
```

```python
#Method Example
class Dog:
    def bark(self):
        return "Woof!"

my_dog = Dog()
print(my_dog.bark())  # Output: Woof!
```

**Q3. What are the different ways to define and call a function in Python?**
- In Python, there are several ways to define and call functions, each serving different purposes. Here are the main methods:

*Standard Function Definition*

The most common way to define a function is using the def keyword followed by the function name and parentheses

**Example**


```python
def greet():
    print("Hello, World!")

# Calling the function
greet()  # Output: Hello, World!
```

*Function with Parameters*

Functions can be defined with parameters to accept input values.

**Example**

```python
def add(a, b):
    return a + b

# Calling the function with arguments
result = add(5, 3)
print(result)  # Output: 8
```

*Function with Default Parameters*

You can provide default values for parameters, allowing the function to be called with fewer arguments.

**Example**

```python
def multiply(a, b=2):
    return a * b

# Calling the function with one argument
print(multiply(4))      # Output: 8 (uses default b=2)
print(multiply(4, 3))   # Output: 12 (b=3 overrides default)
```

*Lambda Functions*

Lambda functions are small anonymous functions defined using the lambda keyword. They are often used for short-term use where defining a full function is unnecessary.

**Example**

```python
square = lambda x: x ** 2

# Calling the lambda function
print(square(5))  # Output: 25
```

*Function Returning Another Function*

Functions in Python can return other functions, which is useful for creating closures.

**Example**

```python
def outer_function(msg):
    def inner_function():
        print(msg)
    return inner_function

# Creating a function that prints a message
my_func = outer_function("Hello from the inner function!")
my_func()  # Output: Hello from the inner function!
```

**Q6. Explain the concept of generators in Python and how they are defined.**
- A generator is defined like a normal function but uses the yield statement instead of return. When a generator function is called, it does not execute its body immediately. Instead, it returns a generator object that can be iterated over to produce values one at a time. Each time the generator's __next__() method is called, the function resumes execution until it hits another yield statement, at which point it pauses again.

**Example**

```python
def count_time(n):
    """Generator that yields numbers from 0 to n-1."""
    value = 0
    while value < n:
        yield value  # Yield current value
        value += 1   # Increment value
#creating a generator
gen=count_time(5)

# Using the generator
for number in gen:
    print(number)  # Output: 0, 1, 2, 3, 4
```

**Q7. What are the advantages of using generators over regular functions?**
- Generators in Python offer several advantages over regular functions

1. Memory Efficiency
2. Lazy Evaluation
3. State Preservation
4. Simpler Code
5. Infinite Sequences

**Example Comparison**


```python
#Regular Function Example
def generate_squares(n):
    squares = []
    for i in range(n):
        squares.append(i * i)
    return squares

# Using the regular function
squares_list = generate_squares(5)
print(squares_list)
```

```python
#Generator Example
def generate_squares_generator(n):
    for i in range(n):
        yield i * i

# Using the generator
for square in generate_squares_generator(5):
    print(square)  # Output: 0, 1, 4, 9, 16
```

**Q10. What is the difference between 'map(), reduce(), and filter() functions in Python?**
- The map(), filter(), and reduce() functions in Python are all higher-order functions that facilitate functional programming by allowing you to apply a function to a sequence of data. However, they serve different purposes and operate in distinct ways.

while all three functions—map(), filter(), and reduce()—are used for processing iterables in Python, they have distinct roles. We will use map() when we want to transform elements, filter() when we need to select elements based on conditions, and reduce() when we want to aggregate elements into a single result.

**Example**


```python
#Use of map to find square of list
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
```

```python
#use of filter to check even numbers in a list
numbers = [1, 2, 3, 4]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4]
```

```python
#Use of reduce for multiplying elements of a list
from functools import reduce

numbers = [2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24 (2 * 3 * 4)
```

```python
# 4.Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    """Check if a number is prime."""
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

user_input = int(input("Enter a number between 1 and 200: "))
if 1 <= user_input <= 200:
  if is_prime(user_input):
    print(f"{user_input} is a prime number.")
  else:
    print(f"{user_input} is not a prime number.")
else:
    print("Please enter a number within the range of 1 to 200.")

```

```python
# 5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms
def fibonacci_sequence(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
if __name__ == "__main__":
    num_terms = 10  # Specify the number of terms you want
    fib_iterator = fibonacci_sequence(num_terms)

    print(f"Fibonacci sequence up to {num_terms} terms:")
    for number in fib_iterator:
        print(number)
```
